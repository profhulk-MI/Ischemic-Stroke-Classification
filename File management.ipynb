```python
import os
import shutil

def copy_images_from_all_subdirs(source_dir, destination_dir):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_dir, exist_ok=True)
    
    total_image_count = 0  # To track total images copied across all directories
    
    for root, dirs, files in os.walk(source_dir):
        # Skip the root directory itself
        if root == source_dir:
            continue
        
        # Get the first-level directory name
        first_level_dir = os.path.basename(os.path.dirname(root))
        
        # Initialize sequence number and count for the current folder
        sequence_number = 1
        folder_image_count = 0
        
        for file in files:
            # Check if the file is an image (adjust extensions as needed)
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
                # New naming format: <first_level_directory_name>_<sequence_number>
                new_name = f"{first_level_dir}_{sequence_number:02d}{os.path.splitext(file)[1]}"
                source_path = os.path.join(root, file)
                destination_path = os.path.join(destination_dir, new_name)
                
                # Copy and rename the file
                shutil.copy(source_path, destination_path)
                folder_image_count += 1
                sequence_number += 1
        
        # Update total image count
        total_image_count += folder_image_count
        
        # Print image count for the current directory
        print(f"Processed directory: {first_level_dir}")
        print(f"Images copied: {folder_image_count}")
    
    # Print total image count
    print(f"\nTotal images copied: {total_image_count}")


# Example usage
source_directory = 'C:/Users/MAHE/Documents/Mahesh/patch importance/Chronic'
destination_directory = 'C:/Users/MAHE/Documents/Mahesh/patch importance/abnormal'
copy_images_with_renaming(source_directory, destination_directory)
```

```python
import os

def count_images_in_directories(source_dir):
    total_images = 0  # To track the total number of images
    
    for root, dirs, files in os.walk(source_dir):
        first_level_dir = os.path.basename(os.path.dirname(root))
        image_count = sum(1 for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')))
            
        total_images += image_count  # Add to the total image count
        print(f"{first_level_dir}: {image_count} images")
    
    # Print the total images count
    print(f"\nTotal images: {total_images}")

# Example usage
source_directory = 'C:/Users/MAHE/Documents/Mahesh/patch importance/Subacute'
count_images_in_directories(source_directory)
```
